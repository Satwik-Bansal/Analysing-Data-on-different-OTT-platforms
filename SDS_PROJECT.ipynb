```python
from scipy.stats import norm
from math import sqrt
#one sided hypothesistest(greater than in alternate hypothesis)
def one_sided_hypo(sample_mean, pop_mean, std_dev, sample_size, alpha):
    actual_z=abs(norm.ppf(alpha))
    hypo_z=(sample_mean-pop_mean)/(std_dev/sqrt(sample_size))
    print('hypothesis z value:',hypo_z)
    print('actual z value:',actual_z,'\n')
    if hypo_z>=actual_z:
        return True
    else:
        return False

alpha=0.05
sample_mean=116
pop_mean=103
sample_size=100
std_dev=20

print('H0 : μ <=',pop_mean)
print('H1 : μ >',pop_mean)
print('alpha value is :',alpha,'\n')

reject=one_sided_hypo(sample_mean, pop_mean, std_dev, sample_size, alpha)
if reject:
    print('Reject NULL hypothesis,since hypothesis z valve is greater than actual z value ')
else:
    print('Failed to reject NULL hypothesis,since hypothesis z value is less than actual value')
```

```python
#Corelation
import matplotlib
import pandas as pd
from matplotlib import pyplot
df=pd.read_csv('movies.csv')
df=df.dropna(axis=0,how='any')
df=df.drop("Type",axis=1)

firstsample=df[1:30]['IMDb']
secondsample=df[1:30]['Age']

pyplot.plot(firstsample,secondsample)
pyplot.show


#Spearman rank correlation
from scipy.stats import spearmanr
stat,p=spearmanr(firstsample,secondsample)

print('stat=%.3f ,p=%5f' % (stat,p))
if p>0.05:
    print('independent variables')
else:
    print('dependent variables')
```
